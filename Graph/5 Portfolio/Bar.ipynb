```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取 Excel 文件
df = pd.read_excel("Portfolio Return Monthly.xlsx")

# 确保列名正确（根据你提供的内容）
months = df['Month']
portfolio_return = df['Return']
index_return = df['Index']

# 设置柱状图的位置
x = months - 1  # 转为从0开始的索引，便于绘图
width = 0.35    # 柱子宽度

# 创建图形
plt.figure(figsize=(14, 6))

# 绘制柱状图
plt.bar(x - width/2, portfolio_return, width, label='Portfolio Return', color='red')
plt.bar(x + width/2, index_return, width, label='Index Return', color='green')

# 在第19和第20期之间画竖线（即 x=19 处，因为 x 是从0开始对应 Month 1）
plt.axvline(x=19 - 0.5, color='black', linestyle='--', linewidth=1.5, label='Market Regime Shift')

# 设置 x 轴刻度
plt.xticks(x, months)

# 添加标签和标题
plt.xlabel('Period (Month)')
plt.ylabel('Return')

plt.legend(loc='upper left')

# 显示网格（可选）
plt.grid(axis='y', linestyle='--', alpha=0.5)

# 展示图形
plt.tight_layout()
plt.show()
```

```python
# 读取 Excel 文件
df = pd.read_excel("Portfolio Return Seasonly.xlsx")

# 提取数据
seasons = df['Season']
portfolio_return = df['Return']
index_return = df['Index']

# 设置柱状图位置（从0开始）
x = seasons - 1  # Season 1 → x=0, ..., Season 7 → x=6
width = 0.35

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制柱状图
plt.bar(x - width/2, portfolio_return, width, label='Portfolio Return', color='red')
plt.bar(x + width/2, index_return, width, label='Index Return', color='green')

# 在第6期和第7期之间画竖线
# 第6期对应 x=5，第7期对应 x=6，中间是 x=5.5
plt.axvline(x=4.5, color='black', linestyle='--', linewidth=1.5, label='Market Regime Shift')

# 设置 x 轴刻度
plt.xticks(x, seasons)
plt.xlabel('Period (Season)')
plt.ylabel('Return')

plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# 读取 Excel 文件
df = pd.read_excel("Portfolio Return HY.xlsx")

# 提取数据
periods = df['HY']          # [1, 2, 3]
portfolio_return = df['Return']
index_return = df['Index']

# 设置柱状图位置（从0开始）
x = periods - 1  # HY=1 → x=0, HY=2 → x=1, HY=3 → x=2
width = 0.35

# 创建图形
plt.figure(figsize=(8, 5))

# 绘制柱状图
plt.bar(x - width/2, portfolio_return, width, label='Portfolio Return', color='red')
plt.bar(x + width/2, index_return, width, label='Index Return', color='green')

# 在第2期和第3期之间画竖线
# 第2期对应 x=1，第3期对应 x=2，中间是 x=1.5
plt.axvline(x=1.5, color='black', linestyle='--', linewidth=1.5, label='Market Regime Shift')

# 设置 x 轴
plt.xticks(x, periods)
plt.xlabel('Period (Half - Year)')
plt.ylabel('Return')

plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
