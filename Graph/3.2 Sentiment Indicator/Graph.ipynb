```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel('.xlsx')

df.index = range(len(df))

df['TradingDate'] = pd.to_datetime(df['TradingDate'])
df.set_index('TradingDate', inplace=True)

fig, ax1 = plt.subplots(figsize=(14, 7))

color1 = 'tab:blue'
ax1.set_xlabel('Date', fontsize=12)
ax1.set_ylabel('Close Price', color=color1, fontsize=12)
ax1.plot(df['ClosePrice'], color=color1, label='Close Price', linewidth=1.5)
ax1.tick_params(axis='y', labelcolor=color1)

ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('Investor Sentiment Index', color=color2, fontsize=12)
ax2.plot(df['B'], color=color2, linestyle='--', label='Sentiment Index', linewidth=1.5)
ax2.tick_params(axis='y', labelcolor=color2)

plt.title('China High-Speed Railway Technology Stock Price and Investor Sentiment Index Over Time', fontsize=14, pad=20)

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left', fontsize=10)

ax1.grid(True, alpha=0.3)

plt.tight_layout()

plt.show()
```
