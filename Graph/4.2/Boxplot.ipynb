```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 读取 Excel 文件
df = pd.read_excel("箱线图.xlsx", sheet_name="Sheet1")

# 删除全 NaN 的行（如有）
df = df.dropna()

# 将数据从宽格式转为长格式（便于 seaborn 绘图）
df_melted = df.melt(var_name='Metric', value_name='Error')

custom_palette = ["#FF4B00", "#00A878"]  # 橙红色和青绿色

# 设置绘图风格
sns.set(style="whitegrid")

# 创建横向箱线图
plt.figure(figsize=(8, 4))
ax = sns.boxplot(x='Error', y='Metric', data=df_melted, orient='h', palette=custom_palette)

# 添加标签和标题
plt.xlabel('Error Value')
plt.ylabel('')


# 可选：调整 x 轴范围（因为数据中有极大异常值，如 RMSE=58.8）
ax.set_xlim(0, 5)  # 例如只显示 0~5 区间，突出主体分布

plt.tight_layout()
plt.show()
```
