```python
import pandas as pd
import matplotlib.pyplot as plt

#  Excel 
df = pd.read_excel("100.xlsx", sheet_name="Sheet1")

# 
df = df.dropna()

# 
plt.figure(figsize=(14, 6))
plt.plot(df['Actual'], label='Actual', color='#1f77b4', linewidth=1.2)
plt.plot(df['Predicted'], label='Predicted', color='#ff7f0e', linewidth=1.2)

# 
plt.xlabel('Time ')
plt.ylabel('Closing Price')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

#  100 
#plt.xlim(0, 200)

plt.tight_layout()
plt.show()
```
